```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pickle import load

file_path = '/home/marcel/ros_ws/analytics_ws/src/performance_monitor/data/from_bag.pkl'
dir_path = '/home/marcel/ros_ws/analytics_ws/src/performance_monitor/data'

with open(os.path.join(dir_path, 'pose_data.csv'), 'r') as f:
    pose_data = pd.read_csv(f)

with open(file_path, 'rb') as f:
    data = load(f)
speed = np.array(data['speed'])
time = np.array(data['time'])
time = time - time[0]
plt.plot(time, speed)
```

```python
x_path = [x['x'] for x in data['path']]
y_path = [y['y'] for y in data['path']]

car_x = pose_data['x_pos']
car_y = pose_data['y_pos']
plt.plot(x_path, y_path, '.r')
plt.plot(car_x, car_y, '-b')
```
