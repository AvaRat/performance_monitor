```python
def moving_avg(poses):
    N = int(len(poses)/1000)
    n = N
    moving_avg = []
    tmp = []
    for i in range(len(poses)):
        tmp.append(poses[i])
        if(i >= N):
            moving_avg.append(np.mean(tmp))
            tmp = []
            N = n+N
    return moving_avg
    

x_avg = moving_avg(x_pos)
y_avg = moving_avg(y_pos)
          
```
